## Optian Optimum number of Clusters For Iris Dataset

### Import Important Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

### Load Our Dataset

```python
iris = datasets.load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data.head()
```

### Explore Our Data

```python
data.info()
```

### Optain Optimum Number Of Clusters Using Elbow Method

```python
cost = []
for i in range(1, 11):
  kmean = KMeans(n_clusters=i)
  kmean.fit(data)
  cost.append(kmean.inertia_)


plt.figure(figsize=(8, 6))
plt.title('Elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('Sum Of Square error distance')
plt.plot(range(1, 11), cost)
plt.show()
```

#### We can see that the elbow has been seen in 3 Clusters and when we increase the number of clusters the sum of square distance doesn't decrease significantly So we Have Choose number of Cluster To be 3 Clusters

### Show cluster Result With Center

```python
kmean = KMeans(n_clusters=3)
y_pred = kmean.fit_predict(data)
plt.figure(figsize=(8, 6))
plt.title('Clusters With Centroids')
plt.scatter(data.values[y_pred == 0, 0], data.values[y_pred == 0, 1], c='red', label = 'Iris-setosa')
plt.scatter(data.values[y_pred == 1, 0], data.values[y_pred == 1, 1], c='blue', label = 'Iris-versicolour')
plt.scatter(data.values[y_pred == 2, 0], data.values[y_pred == 2, 1], c='green', label = 'Iris-virginica')
plt.scatter(kmean.cluster_centers_[:,0], kmean.cluster_centers_[:,1], c='yellow', label = 'Centroids')
plt.legend()
plt.show()
```
